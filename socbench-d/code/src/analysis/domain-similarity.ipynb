# Statistical Analysis of the Benchmark Results

```python
from scipy.stats import f_oneway, tukey_hsd
from benchmark import BenchmarkResultSet
import numpy as np
```

```python
result_files = [open(f'data/socbenchd_{i+1}/results_bge_small_10.json').read() for i in range(3)]
results = [BenchmarkResultSet.model_validate_json(result_file) for result_file in result_files]
```

```python
NUMBER_OF_DOMAINS = 11
def benchmark_result_set_to_recall(result_set: BenchmarkResultSet):
    return [[entry.recall for entry in result_set.results[i].entries] for i in range(NUMBER_OF_DOMAINS)]
results_recall = [benchmark_result_set_to_recall(result_set) for result_set in results]
recalls = []
for i in range(NUMBER_OF_DOMAINS):
    recalls_domain = []
    for j in range(len(results_recall)):
        recalls_domain.extend(results_recall[j][i])
    recalls.append(recalls_domain)
```

Perform ANOVA

```python
f_oneway(*recalls)
```

Check group similarity

```python
result = tukey_hsd(*recalls)
print(result)
```
